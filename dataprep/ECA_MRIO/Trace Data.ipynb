```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
df = pd.read_csv('C://Users/dadel/Downloads/data_trace.csv')
```

```python
df
```

```python
df[df["original_inventory_sector"]=="steel"]
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], format='%m-%d-%Y', errors='coerce')

# Filter rows where gas is 'co2e_20yr'
df_filtered = df[df['gas'] == 'co2e_20yr']

# Sort by source_id and start_time to ensure the latest date is kept
df_sorted = df_filtered.sort_values(by=['source_id', 'start_time'], ascending=[True, False])

# Keep only the row with the latest start_time for each source_id
df_latest = df_sorted.drop_duplicates(subset='source_id', keep='first')

df_latest = df_latest.dropna(subset=['lat', 'lon'])
```

```python
df_latest[df_latest["original_inventory_sector"]=="petrochemicals"]
```

```python
gdf = gpd.GeoDataFrame(df_latest, 
                       geometry=gpd.points_from_xy(df_latest['lon'], df_latest['lat']),
                       crs="EPSG:4326")  # Assuming lat/lon are in WGS84 (EPSG:4326)

# Plotting the map with a larger figure size
fig, ax = plt.subplots(figsize=(15, 10))  # Adjust the figure size here

# Plot the points, scaling by 'capacity_factor' and coloring by 'original_inventory_sector'
gdf.plot(ax=ax, 
         markersize=gdf['capacity_factor'] * 200,  # Adjust scaling factor as necessary
         column='original_inventory_sector', 
         legend=True, 
         cmap='tab20',  # You can change the colormap
         alpha=0.6)  # Adjust transparency of points

# Add basemap (use OpenStreetMap)
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, crs=gdf.crs.to_string())

# Move the legend to the side
legend = ax.get_legend()
legend.set_bbox_to_anchor((1, 1))  # Move the legend outside the plot
legend.set_title("Sector")  # Optional: change legend title

# Labeling the plot
plt.title('Map of Emissions by Capacity Factor and Inventory Sector', fontsize=15)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

plt.show()
```

```python
# Filter out rows with missing or invalid 'lon', 'lat', or 'capacity_factor' values
df_filtered = df_latest.dropna(subset=['lon', 'lat', 'capacity_factor'])
df_filtered = df_filtered[df_filtered['capacity_factor'] > 0]  # Ensure capacity_factor is positive

# Create GeoDataFrame with filtered data
gdf = gpd.GeoDataFrame(df_filtered, 
                       geometry=gpd.points_from_xy(df_filtered['lon'], df_filtered['lat']),
                       crs="EPSG:4326")  # Assuming lat/lon are in WGS84 (EPSG:4326)

# Define markers for each unique sector
markers = ['o', 's', '^', 'D', 'v', '*', 'P', 'X']  # Different shapes for points
unique_sectors = gdf['original_inventory_sector'].dropna().unique()  # Get unique sectors, skipping NaN values
sector_markers = {sector: markers[i % len(markers)] for i, sector in enumerate(unique_sectors)}  # Map each sector to a marker

# Plotting the map with a larger figure size
fig, ax = plt.subplots(figsize=(15, 10))  # Adjust the figure size here

# Plot each sector with a different marker, skipping empty subsets
for sector, marker in sector_markers.items():
    subset = gdf[gdf['original_inventory_sector'] == sector]  # Filter for each sector
    if not subset.empty:  # Only plot if the subset has data
        subset.plot(ax=ax, 
                    markersize=subset['capacity_factor'] * 200,  # Scale by 'capacity_factor'
                    marker=marker,  # Use specific marker for the sector
                    label=sector,  # Label for legend
                    color='blue',  # Set color if not using colormap
                    alpha=0.6)  # Adjust transparency of points

# Add basemap (use OpenStreetMap)
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, crs=gdf.crs.to_string())

# Move the legend to the side
legend = ax.legend(loc="upper left", bbox_to_anchor=(1, 1), title="Sector")  # Move the legend outside the plot

# Labeling the plot
plt.title('Map of Emissions by Capacity Factor and Inventory Sector', fontsize=15)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

plt.show()
```

```python
import folium
import geopandas as gpd
from folium.plugins import MarkerCluster

# Assuming df_latest is your DataFrame with lat, lon, and other required columns

# Initialize a map centered around the midpoint of the data (you can adjust the location)
m = folium.Map(location=[df_latest['lat'].mean(), df_latest['lon'].mean()], zoom_start=5)

# Create a marker cluster for better performance with many points
marker_cluster = MarkerCluster().add_to(m)

# Add points to the map, scaled by 'capacity_factor' and colored by 'original_inventory_sector'
for idx, row in df_latest.iterrows():
    # Create a popup with information about the point
    popup_info = f"Sector: {row['original_inventory_sector']}<br>Capacity Factor: {row['capacity_factor']}"
    
    # Add circle markers with color and radius proportional to the capacity_factor
    folium.CircleMarker(
        location=(row['lat'], row['lon']),
        radius=row['capacity_factor'] * 5,  # Adjust the size factor as necessary
        color=None,
        fill=True,
        fill_color="blue",  # You can assign color based on the sector as needed
        fill_opacity=0.6,
        popup=popup_info
    ).add_to(marker_cluster)

# Save the map to an HTML file or display it directly
m.save('interactive_map.html')
m  # Display the map in a Jupyter notebook or interactive environment
```
